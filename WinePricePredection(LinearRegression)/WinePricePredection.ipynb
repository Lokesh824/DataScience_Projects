### We are not getting any valueable information from the above graph, So Training with linear regression and then we will apply backward elimination to remove unwanted dependent variables 

```python
X=data.iloc[:,1:].values 
```

```python
y=data.iloc[:,1].values
```

### Now we have Dependent Variable (Price) in y and all the independent variables in X array

#### Creating linear regression model and fitting x and y in that 

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X,y)
```

#### We have the test data in differnent csv file so importing the test data to predict the values and check the accurcy 

```python
testData = pd.read_csv(r'D:\MachineLearning\DataScience-Projects\WinePricePredection(LinearRegression)\wine_test.csv')
```

```python
testData.head()
```

```python
X_test = testData.iloc[:,1:].values
y_test = testData.iloc[:,1].values 
```


#### Now use the regressor object to predict the values for the test data and check the accuracy of the model

```python
regressor.predict(X_test)
```

```python
regressor.score(X,y)
```

#### Here we are getting the score of 1.0 means that our model in 100% accurate with the given data
